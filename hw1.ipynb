<h3>Задание 1</h3>

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
(
    ratings[ ratings.rating == 5.0 ][['movieId', 'rating']]
    .groupby('movieId').count().sort_values('rating', ascending = False)
    .merge(movies[['movieId', 'title']], on = 'movieId', how = 'left')
    [['movieId', 'title', 'rating']]
)[0:1]
```

<h3>Задание 2</h3>

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
summed = (
    power[ (power.quantity >= 0) & (power.year >= 2005) & (power.year <= 2010) & 
          (power['country'].isin(['Lithuania', 'Latvia', 'Estonia'])) &
          (power['category'].isin([4, 12, 21]))]
    [['country', 'quantity']]
)

summed['region'] = 'Baltic'

summed.groupby('region').sum()
```
