<h3>Задание 1</h3>

```python
import pandas as pd
```

```python
log = pd.read_csv("visit_log.csv", sep = ";")

log.head()
```

```python
def define_source(row):
    """Разметка источника"""
    
    if row.traffic_source in ('yandex', 'google'):
        return 'organic'
    elif row.region == 'Russia' and row.traffic_source in ('paid', 'email'):
        return 'ad'
    elif row.region != 'Russia' and row.traffic_source in ('paid', 'email'):
        return 'other'
    else:
        return row.traffic_source
```

```python
log['source_type'] = log.apply(define_source, axis = 1)

log.head(15)
```

<h3>Задание 3</h3>

```python
ratings = pd.read_csv('ratings_hw2.csv')

ratings.head()
```

```python
grouped = (
    ratings.groupby('userId')['timestamp'].agg(['count', 'min', 'max']).reset_index()
    .rename(columns={'userId': 'userId', 'count': 'rating_count', 'min': 'min_time', 'max': 'max_time'})
)

grouped['lifetime'] = grouped['max_time'].astype('int') - grouped['min_time'].astype('int')

grouped[ grouped.rating_count > 100 ][['lifetime']].mean()
```
