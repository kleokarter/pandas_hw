```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings_hw2.csv')
```

```python
def movie_class(param):
    """Классифакция фильма по оценкам"""
    
    if param <= 2.0:
        return 'Низкий рейтинг'
    elif param > 2.0 and param <= 4.0:
        return 'Средний рейтинг'
    else:
        return 'Высокий рейтинг'
```

```python
ratings['class'] = ratings['rating'].apply(movie_class)
ratings.head()
```

<h3>Задание 2</h3>

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
def define_region(param):
    """Определение региона по городу, указанном в поисковом запросе"""
    
    region_dict = {
        'Центр': ['москва', 'тула', 'ярославль'],
        'Северо-Запад': ['петербург', 'псков', 'мурманск'],
        'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    
    result = ''
    
    for region in region_dict:
        for city in region_dict[region]:
            if city in param:
                result = region
    
    if result == '':
        return 'undefined'
    else:
        return result
```

```python
keywords['region'] = keywords['keyword'].apply(define_region)
keywords.head()
```

```python
keywords.groupby(['region']).count()
```

<h3>Задание 3</h3>

```python
import re
```

```python
years = []

for i in range(1950, 2011):
    years.append(str(i))
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
def define_movie_year(row):
    """Определить год выпуска фильма по названию"""
    
    movie_year = ''
    
    for year in years:
        if year in row['title']:
            movie_year = year
            
    if movie_year == '':
        return '1900'
    else:
        return movie_year
```

```python
movies['year'] = movies.apply(define_movie_year, axis = 1)

movies.head()
```

```python
result = (
    ratings.merge(movies[['movieId', 'year']], on = 'movieId', how = 'left')
    .groupby('year').mean().reset_index()[['year', 'rating']]
    .sort_values('rating', ascending = False)
)

result.head(50)
```
